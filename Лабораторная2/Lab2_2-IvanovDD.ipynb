Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
class Stack: # Определяем стек
    def __init__(self): # Инициализация
        self.items = []
    def isEmpty(self): # Проверка на пустоту
        return self.items == []
    def push(self, item): # Добавить элемент в начало
        self.items.append(item)
    def pop(self): # Удалить элемент из конца
        return self.items.pop()
    def reverse(self): # Инвертировать
        self.items.reverse()
class Deque: # Определяем дек
    def __init__(self): # Инициализация
        self.items = []
    def isEmpty(self): # Проверка на пустоту
        return self.items == []
    def append(self, item): # Добавить в конец
        self.items.append(item)
    def pop(self): # Удалить из конца
        return self.items.pop()
    def appendleft(self, item): # Добавить в начало
        self.items.insert(0, item)
    def popleft(self): # Удалить из начала
        return self.items.pop(0)
    def peek(self): # Вернуть элемент из очереди без удаления
        return self.items[-1]
    def reverse(self): # Инвертировать
        self.items.reverse()
    def __len__(self): # Длина
        return len(self.items)
def ReadFile(filename): # Чтение из файла
    with open(filename, 'r', encoding = "utf-8") as file:
        return [row.strip() for row in file]
def SaveFile(filename, result): # Сохранение в файл
    with open(filename, 'w', encoding = "utf-8") as f:
        if isinstance(result, str):
            f.write(result + "\n")
        else:
            for item in result:
                f.write(str(item) + "\n")
```
