### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())
def GenerateMatrix(m, n, min, max):
    matrix = []
    for i in range(m):
        line = [] # Генерируем строки
        for j in range(n):
            line.append(random.randint(min, max))
        matrix.append(line) # Добавляем строки в матрицу
    return matrix
matrix1 = GenerateMatrix(user_m, user_n, user_min_limit, user_max_limit)
for i in range(len(matrix1)):
    print(matrix1[i])
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол
# После чего срез уменьшается и цикл повторяется
# Время: O(n^2)
matrix2 = copy.deepcopy(matrix1)
start_time = time.time()
for i in range(len(matrix2)): # Проход каждой строки матрицы
    for j in range(len(matrix2[i])):
        min = j # Индекс j принимается за индекс минимального элемента строки
        for k in range(j + 1, len(matrix2[i])): # Проход элементов строки
            if matrix2[i][k] < matrix2[i][min]:
                min = k
        matrix2[i][j], matrix2[i][min] = matrix2[i][min], matrix2[i][j] # Минимальный элемент сдвигается влево
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix2)):
    print(matrix2[i])
```

```python
# Сортировка вставкой
# Время: O(n) - лучший случай, O(n^2) - другие
matrix3 = copy.deepcopy(matrix1)
start_time = time.time()
for i in range(len(matrix3)): # Проход строк
    for j in range(len(matrix3[i])):
        for k in range(j + 1, len(matrix3[i])):
            while matrix3[i][j] > matrix3[i][k]: # Переносим меньшие элементы влево
                matrix3[i][j], matrix3[i][k] = matrix3[i][k], matrix3[i][j]
                k -= 1
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix3)):
    print(matrix3[i])
```

```python
# Сортировка обменом || Сортировка пузырьком
# Время: O(n) - лучший случай, O(n^2) - другие
matrix4 = copy.deepcopy(matrix1)
start_time = time.time()
for i in range(len(matrix4)): # Проход строк
    for j in range(len(matrix4[i])):
        for k in range(len(matrix4[i]) - 1):
            if matrix4[i][k] > matrix4[i][k + 1]: # Сравнение соседних пар
                matrix4[i][k], matrix4[i][k + 1] = matrix4[i][k + 1], matrix4[i][k]
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix4)):
    print(matrix4[i])
```

```python
# Сортировка Шелла
# Время: O(nlog^2(n)) - лучший случай, O(n^2) - худший
matrix5 = copy.deepcopy(matrix1)
start_time = time.time()
for i in range(len(matrix5)): # Проход строк
    l = len(matrix5[i]) // 2 # Шаг
    while l > 0:
        for j in range(len(matrix5[i])):
            for k in range(j + l, len(matrix5[i])):
                if matrix5[i][j] > matrix5[i][k]: # Сравнение элементов с шагом l
                    matrix5[i][j], matrix5[i][k] = matrix5[i][k], matrix5[i][j] 
            l //= 2
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix5)):
    print(matrix5[i])
```

```python
# Быстрая сортировка
# Время: O(nlog(n)) - лучший случай, O(n^2) - худший
matrix6 = copy.deepcopy(matrix1)
def QuickSort(matrix): # Рекурсивная функция
    if len(matrix) <= 1: # Если длина 1 то ничего не нужно сортировать
        return matrix
    higher, equal, lower = [], [], [] # 3 списка
    base = matrix[0] # Опорный элемент
    for i in range(len(matrix)):
        if matrix[i] < base:
            lower.append(matrix[i])
        elif matrix[i] > base:
            higher.append(matrix[i])
        else:
            equal.append(matrix[i])
    return QuickSort(lower) + equal + QuickSort(higher) # Рекурсия
start_time = time.time()
for i in range(len(matrix6)):
    matrix6[i] = QuickSort(matrix6[i])
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix6)):
    print(matrix6[i])
```

```python
# Турнирная сортировка
# Время: O(nlog^2(n))
matrix7 = copy.deepcopy(matrix1)
def Stack(matrix, StackSize, seed): # Обработка куч
    max = seed
    left = (2 * seed) + 1 # Индексы левого и правого края кучи
    right = (2 * seed) + 2
    if left < StackSize and matrix[left] > matrix[max]: # Проверка левого и правого краев
        max = left
    if right < StackSize and matrix[right] > matrix[max]:
        max = right
    if max != seed:
        matrix[seed], matrix[max] = matrix[max], matrix[seed]
        Stack(matrix, StackSize, max) # Рекурсия
def TournamentSort(matrix):
    n = len(matrix)
    for i in range(n, -1, -1):
        Stack(matrix, n, i)
    for i in range(n - 1, 0, -1): # Сортировка
        matrix[i], matrix[0] = matrix[0], matrix[i]
        Stack(matrix, i, 0)
start_time = time.time()
for i in range(len(matrix7)):
    TournamentSort(matrix7[i])
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix7)):
    print(matrix7[i])
```

```python
# Сортировка встроенным методом
matrix8 = copy.deepcopy(matrix1)
start_time = time.time()
for i in range(len(matrix8)):
    matrix8[i].sort()
print("Время = {0} мс".format((time.time() - start_time)*1000))
for i in range(len(matrix8)):
    print(matrix8[i])
```
